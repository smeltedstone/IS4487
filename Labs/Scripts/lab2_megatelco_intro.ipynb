## Clean up the data
Did you notice anything unusual about the "house" amounts?

## Additional Data Preparation

➡️ Assignment Tasks
- Add at least two additional steps to clean or filter the dataset

```python
df_clean.dropna(inplace=True)

# Displaying information about the DataFrame after removing null values
print(df_clean.info())
```

```python
college_mapping = {'zero': 0, 'one': 1}

# Applying the mapping to the 'college' column
df_clean['college'] = df_clean['college'].map(college_mapping)

# Converting the 'college' column to integer type
df_clean['college'] = df_clean['college'].astype(int)

# Display the updated DataFrame
print(df_clean.info())
```

## Data Visualization

➡️ Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
correlation_matrix = df_clean.corr()

# Display the correlation matrix
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='over_15mins_calls_per_month', y='overage', data=df_clean, scatter_kws={'alpha':0.5})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('over_15mins_calls_per_month')
plt.ylabel('overage')

# Display correlation coefficient on the plot
corr_coefficient = df_clean['over_15mins_calls_per_month'].corr(df['overage'])
plt.text(10, 300, f'Correlation = {corr_coefficient:.2f}', fontsize=12, color='red')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='handset_price', y='income', data=df_clean, scatter_kws={'alpha':0.5})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('handset_price')
plt.ylabel('income')

# Display correlation coefficient on the plot
corr_coefficient = df_clean['handset_price'].corr(df_clean['income'])
plt.text(200, 130000, f'Correlation = {corr_coefficient:.2f}', fontsize=12, color='red')

plt.grid(True)
plt.show()
```
