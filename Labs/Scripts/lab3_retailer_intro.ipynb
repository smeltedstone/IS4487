## Prepare Data

➡️ Assignment Tasks
- Convert the Total_Amount into a number (not a string)
- Remove rows with empty (NULL) values
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
# Replace commas in 'product_price' and 'total_amount' columns and convert to float
df['product_price'] = df['product_price'].str.replace('$', '').str.replace(',', '').astype(float)
df['total_amount'] = df['total_amount'].str.replace('$', '').str.replace(',', '').astype(float)

# Convert 'postal_code' to int
df['postal_code'] = df['postal_code'].astype('Int64')

# Check the data types after conversion
print(df.dtypes)

# Remove rows with NULL values
df.dropna(inplace=True)
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
# Calculate the average price per order month
average_price_per_month = df.groupby('order_month')['product_price'].mean()

print("Average price per order month:")
print(average_price_per_month)
```

```python
# Calculate the total revenue for all order months
total_revenue = df['total_amount'].sum()

print("\nTotal revenue for all order months:", total_revenue)
```
